# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dataset


```python
df = pd.read_csv('match_summary.csv')
```

# Scatter_Plot


```python
sns.scatterplot(x='total_fours', y='total_sixes', data=df)
plt.title('Total Fours vs Total Sixes')
plt.xlabel('Total Fours')
plt.ylabel('Total Sixes')
plt.show()
```

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```
